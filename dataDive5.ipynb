```python
df_2020 = pd.read_csv('2020.csv')
df_2020 = df_2020.sort_values('Country name')
df_2020.tail()
```

```python
same_countries = df_2015
dfs = [df_2016, df_2017, df_2018, df_2019, df_2020]
```

```python
year = 2015
for df in dfs:
    same_countries = same_countries.merge(df, how = 'inner', right_on = 'Country', left_on = 'Country', 
                                         suffixes = (year, year + 1))
    year += 1
```

```python
same_countries = same_countries.dropna()
```

```python
same_countries.set_index('Country', inplace = True)
```
