```python
same_countries = same_countries.drop('Yearly Change %', axis = 1)
```

```python
countries = same_countries
countries
```

```python
countries['Region'].value_counts()
```

```python
regions = []
```

```python
for region in countries['Region']:
    if region == "North America":
        regions.append(0)
    elif region == "Latin America and Caribbean":
        regions.append(1)
    elif region == "Western Europe":
        regions.append(2)
    elif region == "Central and East Europe":
        regions.append(3)
    elif region == "Middle East and Northern Africa":
        regions.append(4)
    elif region == "Sub-Saharan Africa":
        regions.append(5)
    elif region == "Southern Asia":
        regions.append(6)
    elif region == "Eastern Asia":
        regions.append(7)
    elif region == "Southeastern Asia":
        regions.append(8)
    else:
        regions.append(9)
```

```python
countries['RegionID'] = regions
```

```python
countries.head()
```

```python
for column in countries.columns.difference(['Region']):
    data=countries.copy()
    data.groupby('RegionID')[column].mean().plot.bar()
    plt.xlabel('Region')
    plt.ylabel(column)
    plt.title(column)
    plt.show()
```

```python
for column in countries.columns.difference(['Region','RegionID']):
    data=countries.copy()
    sns.lmplot(x = column, y = 'Happiness Score2020', data = data)
```
